```python
train_df = pd.merge(train_values,train_labels,on='patient_id')
train_df.shape
train_df[['slope_of_peak_exercise_st_segment', 'thal', 'chest_pain_type',
       'num_major_vessels', 'fasting_blood_sugar_gt_120_mg_per_dl',
       'resting_ekg_results', 'sex', 'exercise_induced_angina']] = train_df[['slope_of_peak_exercise_st_segment', 'thal', 'chest_pain_type',
       'num_major_vessels', 'fasting_blood_sugar_gt_120_mg_per_dl',
       'resting_ekg_results', 'sex', 'exercise_induced_angina']].astype(str)
train_df2 = pd.get_dummies(train_df.set_index('patient_id'))
train_df2.head()

test_values[['slope_of_peak_exercise_st_segment', 'thal', 'chest_pain_type',
       'num_major_vessels', 'fasting_blood_sugar_gt_120_mg_per_dl',
       'resting_ekg_results', 'sex', 'exercise_induced_angina']] = test_values[['slope_of_peak_exercise_st_segment', 'thal', 'chest_pain_type',
       'num_major_vessels', 'fasting_blood_sugar_gt_120_mg_per_dl',
       'resting_ekg_results', 'sex', 'exercise_induced_angina']].astype(str)
test_values2 = pd.get_dummies(test_values.set_index('patient_id'))
```

```python
X = train_df2.drop('heart_disease_present',axis=1)
y = train_df2[['heart_disease_present']]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = .3, stratify =y,random_state =123)
```

```python
# Import Lasso
from sklearn.linear_model import Lasso

# Instantiate a lasso regressor: lasso
lasso = Lasso(alpha = 0.001,normalize =True)

# Fit the regressor to the data
lasso.fit(X,y)

# Compute and print the coefficients
lasso_coef = lasso.coef_
print(lasso_coef)

# Plot the coefficients
plt.plot(range(len(X.columns)), lasso_coef)
plt.xticks(range(len(X.columns)), X.columns.values, rotation=90)
plt.margins(0.02)
plt.show()
```

```python
lasso_features = pd.Series(dict(zip(X.columns,abs(lasso_coef)))).sort_values(ascending =False)
```

## Random forest feature importance

```python
rf = RandomForestClassifier(n_estimators=101,criterion='gini',max_features='log2')
rf.fit(X,y)
features=X.columns
importances = rf.feature_importances_
indices = np.argsort(importances)

plt.figure(1)
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), features[indices])
plt.xlabel('Relative Importance')
```

```python
feature_list = []
accuracy_list =[]
oob_score = []
rf_feat =  RandomForestClassifier(n_estimators=101,criterion='gini',max_features='log2',oob_score=True)

for element in random_forest_importances.index:
    feature_list.append(element)
    rf_feat.fit(X_train[feature_list],y_train)
    y_pred = rf_feat.predict(X_test[feature_list])
    accuracy_list.append(accuracy_score(y_test,y_pred))
    oob_score.append(rf_feat.oob_score_)
    
print(accuracy_list)
print(oob_score)
```

```python
from sklearn import metrics
rf_feat.fit(X_train[['age',
 'chest_pain_type_4',
 'max_heart_rate_achieved',
 'num_major_vessels_0',
 'oldpeak_eq_st_depression',
 'resting_ekg_results_1',
 'sex_0',
 'thal_normal',
 'thal_reversible_defect']],y_train)
y_pred_proba = rf_feat.predict_proba(X_test[['age',
 'chest_pain_type_4',
 'max_heart_rate_achieved',
 'num_major_vessels_0',
 'oldpeak_eq_st_depression',
 'resting_ekg_results_1',
 'sex_0',
 'thal_normal',
 'thal_reversible_defect']])[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
